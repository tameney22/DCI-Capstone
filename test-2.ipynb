```python
import pandas as pd

df = pd.read_excel("19thCenturyUSNewspapers.xlsx")
df.dropna(inplace=True) # dropped metadata rows

df.head()
```

```python
import os

xmlDF = pd.DataFrame()
for filename in os.listdir("XML/5AHK/NCNP0028/"):
  if filename.endswith("Issue.xml"):
    continue
  issueFilename = filename.replace("Text", "Issue")
  # print(filename)
  row = (df[df["Filename"] == issueFilename][["PublicationTitle", "DataLocation", "Filename"]])
  xmlDF = pd.concat([xmlDF, row], ignore_index=True)
  # break
  
xmlDF.head()
```

```python
xmlDF["Location"] = xmlDF["DataLocation"].str.replace(r"\\19thCenturyUSNewspapers_01\\XML\\NEWSPAPERS", "XML", regex=True) + xmlDF["Filename"].str.replace("Issue", "Text", regex=True)
xmlDF["Location"] = xmlDF["Location"].str.replace(r"\\", "/", regex=True)

xmlDF.drop(['DataLocation', 'Filename'], axis=1, inplace=True)

xmlDF.shape
xmlDF.to_csv("sampleXMLs.csv")
```
